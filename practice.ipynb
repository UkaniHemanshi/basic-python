```python
#line chart
import matplotlib.pyplot as plt
x=[2,8,1,4,6,2,1,9]
plt.plot(range(1,9),x)
plt.show()
```

```python
#multiline chart
import matplotlib.pyplot as plt
x=[2,9,3,6,0,1,5]
y=[4,8,2,5,6,1,3]
plt.plot(range(1,8),x)
plt.plot(range(1,8),y)
#plt.show()
plt.savefig('ex.png',format='png')
```
